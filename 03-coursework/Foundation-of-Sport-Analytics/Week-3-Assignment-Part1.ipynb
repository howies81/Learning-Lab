## Part 1 - NBA Player Comparision

```python
# As usual, we begin by importing the packages we will need

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The data consists of the shot log for the NBA season 2016/17

shot = pd.read_csv("Assignment Data/Week 3/NBA shotlog_16_17.csv")
pd.set_option('display.max_columns', 100)
print(shot.columns.tolist())
shot.describe()
```

```python
#Use the same code that we used to project all shots onto a half court 
# by defining shot[‘halfcourt_x’] and shot[‘halfcourt_y’]

shot['halfcourt_x'] =np.where(shot['location_x'] < 933/2, 933 - shot['location_x'],shot['location_x'])
shot['halfcourt_y'] =np.where(shot['location_x'] < 933/2, 500 - shot['location_y'],shot['location_y'])
shot.describe()
```

## Westbrook - Durant Comparision

```python
# Player's stats
R_Westbrook = shot[shot['shoot_player']=='Russell Westbrook']
K_Durant = shot[shot['shoot_player']=='Kevin Durant']
D_Howard = shot[shot['shoot_player']=='Dwight Howard']
D_Jordan = shot[shot['shoot_player']=='DeAndre Jordan']
B_Lopez = shot[shot['shoot_player']=='Brook Lopez']
R_Lopez = shot[shot['shoot_player']=='Robin Lopez']

# Westbrook - Durant Comparision
hx_R_Westbrook = R_Westbrook['halfcourt_x']
hy_R_Westbrook = R_Westbrook['halfcourt_y']
hx_K_Durant = K_Durant['halfcourt_x']
hy_K_Durant = K_Durant['halfcourt_y']

f = plt.figure(figsize=(94/6,50/6))
ax = f.add_subplot(121)
colors = np.where(R_Westbrook['current_shot_outcome']=='SCORED','r',np.where(R_Westbrook['current_shot_outcome']=='MISSED','b','g'))
ax = plt.scatter(hx_R_Westbrook,hy_R_Westbrook, s=10, c= colors, marker= '.')
plt.grid(True)
plt.title("Russell Westbrook", fontsize = 15)
ax = f.add_subplot(122)
colors = np.where(K_Durant['current_shot_outcome']=='SCORED','r',np.where(K_Durant['current_shot_outcome']=='MISSED','b','g'))
ax = plt.scatter(hx_K_Durant,hy_K_Durant, s=10, c= colors, marker= '.')
plt.grid(True)
plt.title("Kevin Durant", fontsize = 15)
```

## Jordan - Howard Comparision

```python
# Jordan - Howard Comparision
hx_D_Howard = D_Howard['halfcourt_x']
hy_D_Howard = D_Howard['halfcourt_y']
hx_D_Jordan = D_Jordan['halfcourt_x']
hy_D_Jordan = D_Jordan['halfcourt_y']



f = plt.figure(figsize=(94/6,50/6))
ax = f.add_subplot(121)
colors = np.where(D_Howard['current_shot_outcome']=='SCORED','r',np.where(D_Howard['current_shot_outcome']=='MISSED','b','g'))
ax = plt.scatter(hx_D_Howard,hy_D_Howard, s=10, c= colors, marker= '.')
plt.grid(True)
plt.xlim(500,950)
plt.ylim(0,500)
plt.title("Dwight Howard", fontsize = 15)
ax = f.add_subplot(122)
colors = np.where(D_Jordan['current_shot_outcome']=='SCORED','r',np.where(D_Jordan['current_shot_outcome']=='MISSED','b','g'))
ax = plt.scatter(hx_D_Jordan,hy_D_Jordan, s=10, c= colors, marker= '.')
plt.grid(True)
plt.xlim(500,950)
plt.ylim(0,500)
plt.title("DeAndre Jordan", fontsize = 15)
```
