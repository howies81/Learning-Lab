```python
# As usual, we begin by loading the packages we will need

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
# Now we load the data

IPL=pd.read_excel("Assignment Data/Week 5/IPL (assignment) data.xlsx")
```

```python
IPL.describe()
```

```python
IPL.info()
```

```python
# Create the sum of salaries in each season
Sumsal = IPL.groupby(['year'])['salaries'].sum().reset_index().rename(columns={'salaries':'allsal'})
Sumsal
```

```python
#Create a variable for team salary divided by total salaries for that season (relsal).
IPL = pd.merge(IPL, Sumsal, on=['year'], how='left')
display(IPL)
```

```python
IPL['relsal'] = IPL['salaries']/IPL['allsal']
display(IPL)
```

```python
# Create a value for win percentage. Define win percentage as wins divided games with a result 
# (= games played minus games with no result). 
IPL['games_result'] = IPL['played'] - IPL['noresult']
IPL['win_pct'] = IPL['won']/IPL['games_result']
display(IPL)
```

```python
#Create the lagged value of win percentage for each team

# first we sort the values

IPL.sort_values(by=['team','year'], ascending=True)
```

```python
# this will allow us to inspect all rows in the data

pd.set_option('display.max_rows', 100)
IPL
```

```python
# now we create the lagged dependend variable

IPL['win_pct_lag'] = IPL.groupby('team')['win_pct'].shift(1)
IPL
```

```python
#Regress win percentage on:
#a) Relsal
#b) Relsal + lagged win percentage
#c) Relsal + lagged win percentage  + team fixed effects
```

```python
#a) Relsal
win_pct_sal1_lm = smf.ols(formula = 'win_pct ~ relsal', data=IPL).fit()
print(win_pct_sal1_lm.summary())
```

```python
#b) Relsal + lagged win percentage
win_pct_sal2_lm = smf.ols(formula = 'win_pct ~ relsal + win_pct_lag', data=IPL).fit()
print(win_pct_sal2_lm.summary())
```

```python
#c) Relsal + lagged win percentage  + team fixed effects
win_pct_sal3_lm = smf.ols(formula = 'win_pct ~ relsal + win_pct_lag +C(team)', data=IPL).fit()
print(win_pct_sal3_lm.summary())
```
